```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Load data set
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/city_lifestyle_dataset.csv')
df.head()
```

```python
#Remove non-numeric columns
df1 = df.drop(['city_name', 'country'], axis=1)
```

```python
#Scale the numeric data
scaler = StandardScaler()
df1 = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
```

```python
# Run KMeans with 3 clusters
kmn = KMeans(n_clusters=3, random_state=42)

# Create clusters ONLY (not adding to df1)
Clusters = kmn.fit_predict(df1)    # df1 must be numeric-only

Clusters
```

```python

df1.head()
```

```python
# Create a NEW dataframe df2 that contains df1 + clusters
df2 = df1.copy()
df2['cluster'] = Clusters

# Show the df2 dataframe
df2.head()
```

```python
df2.tail()
```

```python
# Plot points
scatter_plot = plt.scatter(df2['avg_income'], df2['avg_rent'], c=df2['cluster'])
plt.legend(*scatter_plot.legend_elements(), title="Cluster")

# Get centroids
centroids = kmn.cluster_centers_

# Plot centroids using correct column indexes
plt.scatter(
    centroids[:, df2.columns.get_loc('avg_income')],
    centroids[:, df2.columns.get_loc('avg_rent')],
    marker='X', s=300, c='red'
)

plt.xlabel("Average Income")
plt.ylabel("Average Rent")
plt.title("Clusters with Centroids")
plt.show()
```

```python
# New city values
x_raw = pd.DataFrame({
    'population_density': [1500],
    'avg_income': [55000],
    'internet_penetration': [0.85],
    'avg_rent': [900],
    'air_quality_index': [40],
    'public_transport_score': [0.7],
    'happiness_score': [6.0],
    'green_space_ratio': [0.3]
})

# Reorder columns to match df1 exactly
x_raw = x_raw[df1.columns]

# Scale using the SAME scaler
x_scaled = scaler.transform(x_raw)

# Predict the cluster
kmn.predict(x_scaled)
```

```python
inertia_list = []

for k in range(1, 10):
    kmn = KMeans(n_clusters=k, random_state=42)
    kmn.fit(df2)    # MUST BE NUMERIC DATA ONLY
    inertia_list.append(kmn.inertia_)

plt.plot(range(1, 10), inertia_list, 'o-')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```
